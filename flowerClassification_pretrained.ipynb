**Import TensorFlow and other libraries**

```python
import matplotlib.pyplot as plt
import numpy as np
import os
import PIL
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
import pathlib
from keras.models import load_model
```

```python
data_dir = pathlib.Path("./flower_photos")
data_dir
```

```python
batch_size = 32
img_height = 180
img_width = 180
```

```python
print("PRE_TRAINED MODEL IS USED..!!")
#This model gives better accuracy and recognise character 
model = load_model('./flower.h5', compile = True)
```

## **Predict on new data**

```python
imageName = "./flower_photos/tulips/142235237_da662d925c.jpg"
PIL.Image.open(imageName)
```

```python
img = tf.keras.utils.load_img(
    imageName, target_size=(img_height, img_width)
)
img_array = tf.keras.utils.img_to_array(img)
img_array = tf.expand_dims(img_array, 0) # Create a batch
predictions = model.predict(img_array)
score = tf.nn.softmax(predictions)
flowerType = np.argmax(score)
```

```python
flowerName = ['daisy', 'dandelion', 'roses', 'sunflowers', 'tulips']
```

```python
print("Predicted image is :: " + str(flowerName[np.argmax(score)]))
```
